```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
n = int(input("Enter the number of vertices : "))
```

```python
x = []
y = []

for i in range(n):
    x.append(random.random() * 10)
    y.append(random.random() * 10)

plt.scatter(x=x, y=y)
```

```python
points = [(x[i], y[i]) for i in range(n)]
print(points)
```

```python
p0 = (11,11)
for i in points:
    if p0[0] > i[0]:
        p0 = i

print(p0)

points.remove(p0)
```

```python
def polar_angle(pi, pj):
    nu = pi[1] - pj[1]
    de = (((pi[0] - pj[0]) ** 2) + ((pi[1] - pj[1]) ** 2)) ** 0.5
    return np.arccos(nu / de)
```

```python

p = sorted(points, key = lambda x : polar_angle(p0, x))
print(p)
```

```python
stack = []
stack.append(p0)
stack.append(p[0])
stack.append(p[1])
```

```python
def non_left_turn(p, q, r):
    val = (q[1] - p[1]) * (r[0] - q[0]) - (q[0] - p[0]) * (r[1] - q[1])
    return True if val <=0 else False
```

```python
for i in range(2, n - 1):
    while non_left_turn(p[i], stack[-1], stack[-2]):
        stack.pop()
    stack.append(p[i])
```

```python
x_val = [x[0] for x in stack]
y_val = [x[1] for x in stack]

x_val.append(p0[0])
y_val.append(p0[1])

plt.scatter(x=x, y=y)
plt.plot(x_val, y_val, color='red')
```
